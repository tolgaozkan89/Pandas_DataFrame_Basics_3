```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic = titanic.iloc[:50,:]
titanic
```

```python
titanic = titanic.set_index(["pclass","sex"]).sort_index(ascending = True)
```

```python
titanic
```

```python
titanic.loc[1]
```

```python
titanic.loc[[1,2]]
```

```python
titanic.loc[:2]
```

```python
titanic.loc[1, "female"]
```

```python
titanic.loc[1,"female","age"] # gives an error
```

```python
titanic.loc[(1,"female")]
```

```python
titanic.loc[(1,"female"),"age"]
```

```python
titanic.loc[([1,2],"female"),["age","fare"]]
```

```python
titanic.loc[(1,"female"),["age","fare"]]
```

```python
titanic.loc[([1,2],"female"),:]
```
