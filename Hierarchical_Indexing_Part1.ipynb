```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic = titanic.iloc[:50, :]
```

```python
titanic.set_index("pclass")
```

```python
titanic.set_index(["pclass","sex"], inplace= True)
```

```python
titanic.sort_index()
```

```python
titanic.sort_index(ascending = False)
```

```python
titanic.sort_index(ascending = [False, True])
```

```python
titanic.swaplevel()
```

```python
titanic.reset_index(inplace = True)
```
